```python
import pandas as pd
import numpy as np
```

```python
url = 'https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv'
df = pd.read_csv(url)
df
```

```python
df2 = df[['Survived', 'Pclass','Age' , 'Fare']]
df2
```

Результат внутри блокнота

```python
df.groupby(['Pclass']).agg({
    'Fare': 'mean',
    'Survived': 'sum',
    'Age': 'mean'
}).reset_index()
```
